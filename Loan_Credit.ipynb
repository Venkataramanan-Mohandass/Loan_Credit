# Data Sourcing

Data Sourcing has already been done and, presented to us with two data sets (in CSV format), namely the application_data and previous_application, that are refered to as "A_data" and "PA_data" in our Analysis.

**A_data** contains all the information of the client at the time of application, about whether or not a client has payment difficulties.

Normalized columns have mean~=median hence, we use mean for imputation.

```python
#Imputing the missing values with mean, for the Numeric Normalized columns

A_data['FLOORSMAX_AVG'].fillna(A_data['FLOORSMAX_AVG'].mean(), inplace=True)
A_data['FLOORSMAX_MODE'].fillna(A_data['FLOORSMAX_MODE'].mean(), inplace=True)
A_data['FLOORSMAX_MEDI'].fillna(A_data['FLOORSMAX_MEDI'].mean(), inplace=True)
A_data['YEARS_BEGINEXPLUATATION_AVG'].fillna(A_data['YEARS_BEGINEXPLUATATION_AVG'].mean(), inplace=True)
A_data['YEARS_BEGINEXPLUATATION_MODE'].fillna(A_data['YEARS_BEGINEXPLUATATION_MODE'].mean(), inplace=True)
A_data['YEARS_BEGINEXPLUATATION_MEDI'].fillna(A_data['YEARS_BEGINEXPLUATATION_MEDI'].mean(), inplace=True)
A_data['TOTALAREA_MODE'].fillna(A_data['TOTALAREA_MODE'].mean(), inplace=True)
A_data['EXT_SOURCE_3'].fillna(A_data['EXT_SOURCE_3'].mean(), inplace=True)
A_data['EXT_SOURCE_2'].fillna(A_data['EXT_SOURCE_2'].mean(), inplace=True)
```

```python
#Print the missing values, for the Numeric Normalized columns

print('FLOORSMAX_AVG =', A_data['FLOORSMAX_AVG'].isna().sum())
print('FLOORSMAX_MODE =', A_data['FLOORSMAX_MODE'].isna().sum())
print('FLOORSMAX_MEDI =', A_data['FLOORSMAX_MEDI'].isna().sum())
print('YEARS_BEGINEXPLUATATION_AVG =', A_data['YEARS_BEGINEXPLUATATION_AVG'].isna().sum())
print('YEARS_BEGINEXPLUATATION_MODE =', A_data['YEARS_BEGINEXPLUATATION_MODE'].isna().sum())
print('YEARS_BEGINEXPLUATATION_MEDI =', A_data['YEARS_BEGINEXPLUATATION_MEDI'].isna().sum())
print('TOTALAREA_MODE =', A_data['TOTALAREA_MODE'].isna().sum())
print('EXT_SOURCE_3 =', A_data['EXT_SOURCE_3'].isna().sum())
print('EXT_SOURCE_2 =', A_data['EXT_SOURCE_2'].isna().sum())
```

We could see that there are now no missing values for any of the Numeric Normalized columns.

These Numeric columns have discrete values i.e., cannot have decimal values. Hence, we use mode for imputation.

```python
#Imputing the missing values with mode, for the Numeric Discrete columns

A_data['AMT_REQ_CREDIT_BUREAU_YEAR'] = A_data['AMT_REQ_CREDIT_BUREAU_YEAR'].mode().iloc[0]
A_data['AMT_REQ_CREDIT_BUREAU_HOUR'] = A_data['AMT_REQ_CREDIT_BUREAU_HOUR'].mode().iloc[0]
A_data['AMT_REQ_CREDIT_BUREAU_DAY'] = A_data['AMT_REQ_CREDIT_BUREAU_DAY'].mode().iloc[0]
A_data['AMT_REQ_CREDIT_BUREAU_WEEK'] = A_data['AMT_REQ_CREDIT_BUREAU_WEEK'].mode().iloc[0]
A_data['AMT_REQ_CREDIT_BUREAU_MON'] = A_data['AMT_REQ_CREDIT_BUREAU_MON'].mode().iloc[0]
A_data['AMT_REQ_CREDIT_BUREAU_QRT'] = A_data['AMT_REQ_CREDIT_BUREAU_QRT'].mode().iloc[0]
A_data['DEF_30_CNT_SOCIAL_CIRCLE'] = A_data['DEF_30_CNT_SOCIAL_CIRCLE'].mode().iloc[0]
A_data['OBS_60_CNT_SOCIAL_CIRCLE'] = A_data['OBS_60_CNT_SOCIAL_CIRCLE'].mode().iloc[0]
A_data['DEF_60_CNT_SOCIAL_CIRCLE'] = A_data['DEF_60_CNT_SOCIAL_CIRCLE'].mode().iloc[0]
A_data['OBS_30_CNT_SOCIAL_CIRCLE'] = A_data['OBS_30_CNT_SOCIAL_CIRCLE'].mode().iloc[0]
A_data['CNT_FAM_MEMBERS'] = A_data['CNT_FAM_MEMBERS'].mode().iloc[0]
A_data['DAYS_LAST_PHONE_CHANGE'] = A_data['DAYS_LAST_PHONE_CHANGE'].mode().iloc[0]
```

```python
#Printing the missing values, for the Numeric Discrete columns

print('AMT_REQ_CREDIT_BUREAU_YEAR =', A_data['AMT_REQ_CREDIT_BUREAU_YEAR'].isna().sum())
print('AMT_REQ_CREDIT_BUREAU_HOUR =', A_data['AMT_REQ_CREDIT_BUREAU_HOUR'].isna().sum())
print('AMT_REQ_CREDIT_BUREAU_DAY =', A_data['AMT_REQ_CREDIT_BUREAU_DAY'].isna().sum())
print('AMT_REQ_CREDIT_BUREAU_WEEK =', A_data['AMT_REQ_CREDIT_BUREAU_WEEK'].isna().sum())
print('AMT_REQ_CREDIT_BUREAU_MON =', A_data['AMT_REQ_CREDIT_BUREAU_MON'].isna().sum())
print('AMT_REQ_CREDIT_BUREAU_QRT =', A_data['AMT_REQ_CREDIT_BUREAU_QRT'].isna().sum())
print('DEF_30_CNT_SOCIAL_CIRCLE =', A_data['DEF_30_CNT_SOCIAL_CIRCLE'].isna().sum())
print('OBS_60_CNT_SOCIAL_CIRCLE =', A_data['OBS_60_CNT_SOCIAL_CIRCLE'].isna().sum())
print('DEF_60_CNT_SOCIAL_CIRCLE =', A_data['DEF_60_CNT_SOCIAL_CIRCLE'].isna().sum())
print('OBS_30_CNT_SOCIAL_CIRCLE =', A_data['OBS_30_CNT_SOCIAL_CIRCLE'].isna().sum())
print('CNT_FAM_MEMBERS =', A_data['CNT_FAM_MEMBERS'].isna().sum())
print('DAYS_LAST_PHONE_CHANGE =', A_data['DAYS_LAST_PHONE_CHANGE'].isna().sum())
```

### **Finding out the data imbalance**

```python
#Finding out the imbalance ratio using Pie chart

count1 = 0 
count0 = 0
for i in A_data['TARGET'].values:
    if i == 1:
        count1 += 1
    else:
        count0 += 1
        
count1 = (count1/len(A_data['TARGET']))*100
count0 = (count0/len(A_data['TARGET']))*100

x = ['Defaulted Population(TARGET=1)','Non-Defauted Population(TARGET=0)']
y = [count1, count0]

explode = (0.1, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(y, explode=explode, labels=x, autopct='%1.1f%%',
        shadow=True, startangle=150)
ax1.axis('equal')  
plt.title('Data Imbalance',fontsize=15)
plt.show()
```

<font color='blue'>

**Insights**
    
We can see that the Application data has high imbalance with Defaulted population at 8.1% as compared to Non-defualted population at 91.9%, with an Imbalance ratio of 11.34.

</font>

### Multivariate Analysis

Visualizing the the number of loans that has been refused, approved or cancelled previously, for which the applicant has now paid the dues on time or not, in the form of a graph.

```python
#Function to plot countplot for categorical variables

def plot_func(var):
    fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(15,5))
    
    s1=sns.countplot(ax=ax1,x=refused[var], data=refused, order= refused[var].value_counts().index,)
    ax1.set_title("Refused", fontsize=10)
    ax1.set_xlabel('%s' %var)
    ax1.set_ylabel("Count of Loans")
    s1.set_xticklabels(s1.get_xticklabels(),rotation=90)
    
    s2=sns.countplot(ax=ax2,x=approved[var], data=approved, order= approved[var].value_counts().index,)
    s2.set_xticklabels(s2.get_xticklabels(),rotation=90)
    ax2.set_xlabel('%s' %var)
    ax2.set_ylabel("")
    ax2.set_title("Approved", fontsize=10)
    
    
    s3=sns.countplot(ax=ax3,x=canceled[var], data=canceled, order= canceled[var].value_counts().index,)
    ax3.set_title("Canceled", fontsize=10)
    ax3.set_xlabel('%s' %var)
    ax3.set_ylabel("")
    s3.set_xticklabels(s3.get_xticklabels(),rotation=90)
    plt.show()
```

```python
#Applying our target variable to the countplot function

plot_func('TARGET')
```
